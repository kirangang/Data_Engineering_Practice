How to use explode

explode() is a PySpark function used to convert each element of an array or each key-value pair of a map into separate rows.


```python
#using explode() to separte out skill for each emp_id by creating temporary column 'skill' using alias 
df.select("emp_id",explode(split(col("skills"),',')).alias("skill")).show()
```

```python
df.select(
    "emp_id",
    explode(split(col("skills"), ",")).alias("skill")
).filter(col("skill") == "Python").show()
```

```python
df=df.withColumn("skill",
                 explode(split(col("skills"), ",")).alias("skill")
).filter(col("skill") == "SQL").show()
```
