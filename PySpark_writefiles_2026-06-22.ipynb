### how to write dataframe into catalog or Cloud

```python
student_detail=spark.read.csv('/Workspace/Users/eshitam230@gmail.com/Data_Engineering_Practice/Data_files/student_detail.csv',header=True,inferSchema=True)
student_detail.display()
```

```python
student_detail.write.format("parquet").mode("overwrite").save('/Volumes/workspace/default/my_volume/final_output/')
```

alternative ways to write for the different file format

```python

student_detail.write.csv('/Volumes/workspace/default/my_volume/final_output/csv/',mode='overwrite',header=True)
```

```python
# this is how we write for compression using gzip
#GZIP compresses this content to reduce file size. You cannot open it directly in a text editor without decompressing it.
student_detail.write.option('format','csv').option('mode','overwrite').option('header','true').option('compression','gzip').save('/Volumes/workspace/default/my_volume/final_output/zip_folder')
```

```python
# This tells Spark to use | (pipe) as the column separator instead of the default comma.
student_detail.write.format('csv').option('header','true').option('sep','|').mode('overwrite').save('/Volumes/workspace/default/my_volume/final_output/csv_sep')
```

```python
student_detail.display()
```
