### How to add the column to existing dataframe using withColumn and lit

```python
from pyspark.sql.functions import *
```

```python
#lit adds the constant value
df=df.withColumn('Country',lit('India'))
```

```python
df.display()
```

### How to increment salary with hardcoded value

```python
increment_salary=df.withColumn('new_salary',col('salary')+5000)
```

```python
increment_salary.display()
```

```python
bonus_salary_emp=increment_salary.withColumn('bonus',col('salary')* 0.20)
```

```python
bonus_salary_emp.display()
```

```python
annual_salary=bonus_salary_emp.withColumn('annual_salary',col('new_salary')*12)
```

```python
annual_salary.display()
```

```python
annual_salary=annual_salary.withColumn('future_age',col('age')+5)
```

```python
annual_salary.display()
```

```python
emp_upper_case_name=annual_salary.withColumn('upper_case',upper(col('full_name')))
```

```python
emp_upper_case_name.limit(10).display()
```

### How to use when and otherwise

```python
emp_upper_case_name=emp_upper_case_name.withColumn('salary_grade', when(col('salary')>= 70000,'A').otherwise('B'))
```

```python
emp_upper_case_name.limit(5).display()
```

### Email domain using split

```python
emp_upper_case_name=emp_upper_case_name.withColumn('domain',split(col('email'), '@').getItem(1))
emp_upper_case_name.limit(5).display()
```

```python
emp_upper_case_name=emp_upper_case_name.withColumn('joining_year',split(col('joining_date'), '-').getItem(0))
emp_upper_case_name.limit(5).display()
```
