## how to define the schema manually using structType

```python
from pyspark.sql.types import * 
```

```python
help(StructType)
```

```python
help(spark.createDataFrame)
```

```python
data =[(1,'Rahul'),(2,'Sachin')]
schema=StructType([
    StructField('id',IntegerType(),True), #True means null values accepted
    StructField('name',StringType(),True)   
])
df=spark.createDataFrame(data,schema)
df.display()
```

```python
df.printSchema()
```
